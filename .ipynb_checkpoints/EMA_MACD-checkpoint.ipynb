```python
def backTest_EMA_MACD(df, ticker, investment=10000):
    rsi = df['RSI']
    close = df['Close']
    open = df['Open']
    ema200 = df['EMA200']
    ema50 = df['EMA50']
    macdLine = df['macd']
    macdSignal = df['macdsignal']
    
    date = df['Date']
    
    # initial condition
    in_position = False
    equity = investment
    commission = 50
    share = 0
    entry = []
    exit = []
    tradeCount=0
    buyDate = ''
    sellDate = ''
    buyPrice = 0
    sellPrice = 0

# 50EMA > 200EMA
# Close > 50EMA
# RSI(3) > 80
# MACD_Line > MACD_Signal
# Two bullish Candle with CurrentClose > PrevClose
    
    for i in range(2, len(df)-1):
        if not in_position and (rsi[i] > 80 and (close[i] > ema50[i]) and (ema50[i] > ema200[i]) and (macdLine[i] > macdSignal[i]) \
        and (open[i] < close[i]) and (close[i] > close[i-1]) ):
            share = math.floor(equity / open[i+1] )
            equity -= share * open[i+1]
            buyDate = date[i+1]
            entry.append((i, open[i+1], buyDate))
            in_position = True
            tradeCount += 1
            
            #print(f'Buy {share} shares at {round(close[i],2)} on {df.index[i].strftime("%Y/%m/%d")}')
        elif in_position and (close[i] < ema50[i] or ema50[i] < ema200[i] ) :
                equity += share * open[i + 1] - commission
                sellDate = date[i  + 1]
                exit.append((i + 1, open[i +1], sellDate, equity))
                in_position = False
                
            
            #print(f'Sell at {round(close[i],2)} on {df.index[i].strftime("%Y/%m/%d")}')
    if in_position:
        equity += share * close[i] - commission 
    
    earning = equity - investment
    roi = round(earning/investment*100,2)
    print(f'{ticker} :\t Earning from investing {investment} is ${round(earning,2)} \t (ROI = {roi}%) \t Trades={tradeCount}')
    return entry, exit, earning, roi, tradeCount
```

```python
def NSEData(tickerList, startDate, interval='1D'):
    

    #Fetch NSE Ticker Data
    result=[] 
    for ticker in tickerList: 
        data = []
        symbol = ticker + ".NS" 
        df = yf.download(symbol, start=startDate, interval=interval)
        df['RSI'] = ta.RSI(df['Close'], timeperiod=3)
        df['EMA200'] = ta.EMA(df['Close'], timeperiod=200)
        df['EMA50'] = ta.EMA(df['Close'], timeperiod=50)
        df['macd'],df['macdsignal'], df['macdhist'] = ta.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
        df['Date'] = df.index
        df = df.dropna()
        data.append(symbol)
        data.append(df)
        result.append(data)

    return result
```

```python
# 2. Generate Calls

ROI =[]
Earning =[]
tickerList =[]
investment = []
totalTrades = []
finalResults = []

for data in nseDF:
    df=data[1]
    ticker=data[0]
    entry, exit, earning, roi, tradeCount=backTest_EMA_MACD(df,ticker, 10000)
    ROI.append(roi)
    Earning.append(earning)
    investment.append(10000)
    totalTrades.append(tradeCount)
    tickerList.append(ticker)
    entryDF = pd.DataFrame(entry,columns=['Day','BuyPrice','BuyDate'])
    exitDF = pd.DataFrame(exit,columns=['Day','SellPrice','SellDate','Equity'])
    result = pd.merge(entryDF, exitDF, left_index=True, right_index=True)
    result['PnL_Percent'] = (result['SellPrice'] - result['BuyPrice'])/result['BuyPrice']*100
    result['Symbol'] = ticker
    result.drop(['Day_x', 'Day_y'], axis = 1,inplace=True)
    finalResults.append(result)
    
resultDF = pd.DataFrame(
    {'Ticker': tickerList,
     'Investment': investment,
     'FinalReturns': Earning,
     'ROI': ROI,
     'TotalTrades': totalTrades
    })

finalResults = pd.concat(finalResults)
finalResults
finalResults = finalResults[['Symbol','BuyDate', 'BuyPrice', 'SellDate', 'SellPrice','PnL_Percent','Equity']]
finalResults.to_csv('EMA_MACD_SWING_TRADES.csv')
resultDF.to_csv('EMA_MACD_SWING_Final_Result.csv')
```

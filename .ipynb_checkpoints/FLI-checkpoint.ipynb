```python
def getBB(df):
    high = df['High']
    low = df['Low']
    close = df['Close']
    # calculate Bollinger Bands
    upper, middle, lower = ta.BBANDS(
                                    df['Close'], 
                                    timeperiod=21,
                                    # number of non-biased standard deviations from the mean
                                    nbdevup=1,
                                    nbdevdn=1                                )
    df['BBUpper']=upper
    df['MiddleBand']=middle
    df['BBLower']=lower
    return df

def getATR(df):
    # calculate ATR
    high = df['High']
    low = df['Low']
    close = df['Close']
    atr_period=5
    price_diffs = [high - low, 
                   high - close.shift(), 
                   close.shift() - low]
    true_range = pd.concat(price_diffs, axis=1)
    true_range = true_range.abs().max(axis=1)
    # default ATR calculation in supertrend indicator
    atr = true_range.ewm(alpha=1/atr_period,min_periods=atr_period).mean() 
    df['atr']=atr
    return df
```

```python
def getTrendLine(df):
    
    BBUpper = df['BBUpper']
    BBLower = df['BBLower']
    high = df['High']
    low = df['Low']
    close = df['Close']
    atr=df['atr']

    BBSignal=[False] * len(df)
    TrendLine=[0] * len(df)
    iTrend=[-1] * len(df)

    buy = [False] * len(df)
    sell = [False] * len(df)
    TrendLine[0]=low[0]-atr[0]

    for i in range(1,len(df)):

        if(close[i]>BBUpper[i]):
            BBSignal[i]=1
        elif(close[i]<BBLower[i]):
            BBSignal[i]=-1
        else:
            BBSignal[i]=0

        if(BBSignal[i]==1):
            TrendLine[i]=low[i]-atr[i]
            if TrendLine[i]<TrendLine[i-1]:
                TrendLine[i]=TrendLine[i-1]

        if(BBSignal[i]==-1):
            TrendLine[i]=high[i]+atr[i]
            if TrendLine[i]>TrendLine[i-1]:
                TrendLine[i]=TrendLine[i-1]

        if(BBSignal[i]==0):
            TrendLine[i]=TrendLine[i-1]

        iTrend[i]=iTrend[i-1]
        if TrendLine[i]>TrendLine[i-1] :
            iTrend[i]=1
        if TrendLine[i]<TrendLine[i-1] :
            iTrend[i]=-1

    indicatorFrame = pd.DataFrame(list(zip(BBSignal, TrendLine, iTrend)),columns =['BBSignal', 'TrendLine','iTrend',])
    indicatorFrame.index = df.index
    return indicatorFrame
```

```python

def backTestFollowLine(df, ticker, investment=10000):

    # initial condition
    in_position = False
    equity = investment
    commission = 50
    share = 0
    entry = []
    exit = []
    iTrend = df['iTrend']
    high = df['High']
    low = df['Low']
    close = df['Close']
    open = df['Open']
    date = df['Date']
    
    tradeCount=0
    buyDate = ''
    sellDate = ''
    buyPrice = 0
    sellPrice = 0

    for i in range(0, (len(indicatorFrame)-1)):
        #buy:=iTrend[1]==-1 and iTrend==1 ? 1 : na
        #sell:=iTrend[1]==1 and iTrend==-1? 1 : na

        buy = True if iTrend[i-1]==-1 and iTrend[i]==1 else False
        sell = True if iTrend[i-1]==1 and iTrend[i]==-1 else False
        #print(f'Buy : {buy}, Sell : {sell}')

        if not in_position and buy:
            share = math.floor(equity / open[i+1] )
            equity -= share * open[i+1]
            buyDate = date[i+1]
            entry.append((i, open[i+1], buyDate))
            in_position = True
            tradeCount += 1

        elif in_position and sell:
            equity += share * open[i+1] - commission
            sellDate = date[i+1]
            exit.append((i, open[i+1], sellDate, equity))
            in_position = False

    if in_position:
            equity += share * close[i] - commission

    earning = equity - investment
    roi = round(earning/investment*100,2)
    print(f'{ticker} :\t Earning from investing {investment} is ${round(earning,2)} \t (ROI = {roi}%) \t Trades={tradeCount}')
    return entry, exit, earning, tradeCount
```

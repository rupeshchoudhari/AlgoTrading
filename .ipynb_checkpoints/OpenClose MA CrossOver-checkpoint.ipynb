```python
import pandas as pd
import numpy as np
from datetime import datetime
import yfinance as yf
import math
import matplotlib.pyplot as plt
import talib as ta

```

```python
# This strategy is not working on NSE
def backtest_OCC(df, ticker, investment):
    openSeries = df['OpenMA9']
    close = df['Close']
    closeSeries = df['CloseMA9']
    open = df['Open']
    date = df['Date']
    trendSMA = df['SMA200']
    
    # initial condition
    in_position = False
    equity = investment
    commission = 50
    share = 0
    entry = []
    exit = []
    tradeCount=0
    buyDate = ''
    sellDate = ''
    buyPrice = 0
    sellPrice = 0
    
    for i in range(2, len(df)):
        # if not in position & price is on uptrend -> buy
        if not in_position and (openSeries[i] < closeSeries[i] and openSeries[i-1] > closeSeries[i-1] and close[i]>trendSMA[i]):
            share = math.floor(equity / open[i+1] )
            equity -= share * open[i+1]
            buyDate = date[i+1]
            entry.append((i, open[i+1], buyDate, share))
            in_position = True
            tradeCount += 1
            
            #print(f'Buy {share} shares at {round(close[i],2)} on {df.index[i].strftime("%Y/%m/%d")}')
        # if in position & price is not on uptrend -> sell
        elif in_position and (openSeries[i] > closeSeries[i] and openSeries[i-1] < closeSeries[i-1]):
            equity += share * open[i+1] - commission
            sellDate = date[i]
            exit.append((i, open[i+1], sellDate, equity, share))
            in_position = False
            #print(f'Sell at {round(close[i],2)} on {df.index[i].strftime("%Y/%m/%d")}')
        
    
    earning = equity - investment
    roi = round(earning/investment*100,2)
    print(f'{ticker} :\t Earning from investing {investment} is ${round(earning,2)} \t (ROI = {roi}%) \t Trades={tradeCount}')
    return entry, exit, earning, tradeCount
```

```python
symbol = '^NSEI' 
df = yf.download(symbol, start='2015-01-01') 
df['OpenMA9'] = ta.SMA(df['Open'], timeperiod=9)
df['CloseMA9'] = ta.SMA(df['Close'], timeperiod=9)
df['SMA200'] = ta.SMA(df['Close'], timeperiod=200)
df.dropna()
df
```

```python
df['Date'] = df.index
entry, exit, roi, tradeCount= backtest_OCC(df,symbol, 10000)
entryDF = pd.DataFrame(entry,columns=['Day','BuyPrice','BuyDate', 'share'])
exitDF = pd.DataFrame(exit,columns=['Day','SellPrice','SellDate','Equity','share'])
result = pd.merge(entryDF, exitDF, left_index=True, right_index=True)
result['PnL'] = (result['SellPrice'] - result['BuyPrice'])/result['BuyPrice']*100
result
```

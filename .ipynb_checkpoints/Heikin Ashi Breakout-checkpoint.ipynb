```python
# 1. Last foud candles should be red
# 2. Currunt candle should not be green with Low==Open
# 3. Buy when SignalSince < 5 and Close > HA_High[Signal] 
# 4. Exit for profit 4% or 30th Day or Sell Signal
```

```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import datetime
import yfinance as yf
import math
import matplotlib.pyplot as plt
from pandas.tseries.offsets import MonthEnd
from pandas.tseries.offsets import Week
#from mplfinance import candlestick_ohlc
#from mplfinance.original_flavor import candlestick_ohlc
```

```python
def heikin_ashi(df):
    heikin_ashi_df = pd.DataFrame(index=df.index.values, columns=['HA_Open', 'HA_High', 'HA_Low', 'HA_Close'])
    
    heikin_ashi_df['HA_Close'] = (df['Open'] + df['High'] + df['Low'] + df['Close']) / 4
    
    for i in range(len(df)):
        if i == 0:
            heikin_ashi_df.iat[0, 0] = df['Open'].iloc[0]
        else:
            heikin_ashi_df.iat[i, 0] = (heikin_ashi_df.iat[i-1, 0] + heikin_ashi_df.iat[i-1, 3]) / 2
        
    heikin_ashi_df['HA_High'] = heikin_ashi_df.loc[:, ['HA_Open', 'HA_Close']].join(df['High']).max(axis=1)
    
    heikin_ashi_df['HA_Low'] = heikin_ashi_df.loc[:, ['HA_Open', 'HA_Close']].join(df['Low']).min(axis=1)
    
    return heikin_ashi_df
```

```python
def isPullBack(df):
    if (df['HA_Open'][2] > df['HA_Close'][2]) and (df['HA_Open'][1] > df['HA_Close'][1]) and (df['HA_Open'][0] > df['HA_Close'][0]):
        return True;
    else:
        return False;
```

```python
def backtest_HA(df, ticker, investment):
    open = df['Open']
    HA_Open=df['HA_Open']
    HA_Low=df['HA_Low']
    HA_Close=df['HA_Close']
    HA_High=df['HA_High']
    date = df['Date']
    
    # initial condition
    in_position = False
    equity = investment
    commission = 50
    share = 0
    entry = []
    exit = []
    tradeCount=0
    buyDate = ''
    sellDate = ''
    buyPrice = 0
    sellPrice = 0
    
    for i in range(5, len(df)-1):
        # if not in position & price is on uptrend -> buy
        if not in_position and HA_Open[i]==HA_Low[i] and isPullBack(df[i-3:i]):
            share = math.floor(equity / open[i+1] )
            equity -= share * open[i+1]
            buyDate = date[i+1]
            entry.append((i+1, open[i+1], buyDate))
            in_position = True
            tradeCount += 1
            
            #print(f'Buy {share} shares at {round(close[i],2)} on {df.index[i].strftime("%Y/%m/%d")}')
        # if in position & price is not on uptrend -> sell
        elif in_position and HA_Open[i]==HA_High[i]:
            equity += share * open[i+1] - commission
            sellDate = date[i+1]
            exit.append((i+1, open[i+1], sellDate, equity))
            in_position = False
            
            #print(f'Sell at {round(close[i],2)} on {df.index[i].strftime("%Y/%m/%d")}')
    # if still in position -> sell all share 
    if in_position:
        equity += share * open[i] - commission
    
    earning = equity - investment
    roi = round(earning/investment*100,2)
    print(f'{ticker} :\t Earning from investing {investment} is ${round(earning,2)} \t (ROI = {roi}%) \t Trades={tradeCount}')
    return entry, exit, earning, roi, tradeCount
```

```python
#Generate calls for single stock

startDate='2021-01-01'
ticker = "SBIN" + ".NS" 
df = yf.download(ticker, start=startDate) 
df['Date'] = df.index

frames = [df, heikin_ashi(df)]
finalDF = pd.concat(frames,axis=1, join='inner')
finalDF.dropna()

ROI =[]
tickerList =[]
investment = []
totalTrades = []
finalResults = []
Earning = []
entry, exit, earning, roi, tradeCount = backtest_HA(finalDF,ticker, 10000)
Earning.append(earning)
ROI.append(roi)
investment.append(10000)
totalTrades.append(tradeCount)
tickerList.append(ticker)
entryDF = pd.DataFrame(entry,columns=['Day','BuyPrice','BuyDate'])
exitDF = pd.DataFrame(exit,columns=['Day','SellPrice','SellDate','Equity'])
result = pd.merge(entryDF, exitDF, left_index=True, right_index=True)
result['PnL_Percent'] = (result['SellPrice'] - result['BuyPrice'])/result['BuyPrice']*100
result['Symbol'] = ticker = "SBIN"
result.drop(['Day_x', 'Day_y'], axis = 1,inplace=True)
result.to_csv('result\HA_Result.csv')
```

```python
# Generate calls for NSE Data

ROI =[]
tickerList =[]
investment = []
totalTrades = []
finalResults = []
Earning = []

for data in nseDF:
    df=data[1]
    ticker=data[0]
    entry, exit, earning, roi, tradeCount = backtest_HA(df,ticker, 10000)
    Earning.append(earning)
    ROI.append(roi)
    investment.append(10000)
    totalTrades.append(tradeCount)
    tickerList.append(ticker)
    entryDF = pd.DataFrame(entry,columns=['Day','BuyPrice','BuyDate'])
    exitDF = pd.DataFrame(exit,columns=['Day','SellPrice','SellDate','Equity'])
    result = pd.merge(entryDF, exitDF, left_index=True, right_index=True)
    result['PnL_Percent'] = (result['SellPrice'] - result['BuyPrice'])/result['BuyPrice']*100
    result['Symbol'] = ticker
    result.drop(['Day_x', 'Day_y'], axis = 1,inplace=True)
    finalResults.append(result)
    
resultDF = pd.DataFrame(
    {'Ticker': tickerList,
     'Investment': investment,
     'FinalReturns': ROI,
     'TotalTrades': totalTrades
    })
resultDF['FinalAmount'] = resultDF['Investment'] + resultDF['Investment']*resultDF['FinalReturns']

finalResults = pd.concat(finalResults)
finalResults
finalResults = finalResults[['Symbol','BuyDate', 'BuyPrice', 'SellDate', 'SellPrice','PnL_Percent','Equity']]
finalResults.to_csv('result\HA_BackTest_Trades.csv')
resultDF.to_csv('result\HA_Final_Result.csv')
```

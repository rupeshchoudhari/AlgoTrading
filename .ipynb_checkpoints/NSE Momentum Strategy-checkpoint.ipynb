# NSE Momentum Strategy

```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import datetime
import yfinance as yf
import math
import matplotlib.pyplot as plt
from pandas.tseries.offsets import MonthEnd
from pandas.tseries.offsets import Week
```

```python
# Fetch data for give tickerList
tickerList=['IOB', 'NHPC', 'UNIONBANK', 'IDBI', 'PNB', 'IDFCFIRSTB', 'BHEL', 'BANKINDIA', 'BANKBARODA', 'L&TFH', 'ADANIPOWER', 'TATAPOWER', 'NTPC', 'IOC', 'ONGC', 'SAIL', 'ABCAPITAL', 'ASHOKLEY', 'COALINDIA', 'CANBK', 'TATAMTRDVR', 'GAIL', 'NMDC', 'MOTHERSUMI', 'VEDL', 'HINDPETRO', 'TATAMOTORS', 'AMBUJACEM', 'HINDZINC', 'HINDALCO', 'JINDALSTEL', 'CROMPTON', 'SBIN', 'TORNTPOWER', 'LICHSGFIN', 'MARICO', 'BPCL', 'GUJGASLTD', 'IGL', 'LAURUSLABS', 'BHARTIARTL', 'DABUR', 'ICICIPRULI', 'WIPRO', 'CHOLAFIN', 'SUNTV', 'MCDOWELL-N', 'CADILAHC', 'TATACONSUM', 'CONCOR', 'ICICIBANK', 'TVSMOTOR', 'HDFCLIFE', 'BHARATFORG', 'JSWSTEEL', 'SUNPHARMA', 'AXISBANK', 'ADANIPORTS', 'CUMMINSIND', 'BERGEPAINT', 'UPL', 'M&M', 'TRENT', 'GODREJCP', 'CIPLA', 'RAMCOCEM', 'HCLTECH', 'SBILIFE', 'AUBANK', 'INDUSINDBK', 'VOLTAS', 'HAVELLS', 'AUROPHARMA', 'SBICARD', 'TECHM', 'TATASTEEL', 'LUPIN', 'MUTHOOTFIN', 'ADANIGREEN', 'ADANIENT', 'GODREJPROP', 'ATGL', 'INFY', 'GRASIM', 'ADANITRANS', 'SRTRANSFIN', 'HDFCBANK', 'LT', 'TITAN', 'AARTIIND', 'PEL', 'KOTAKBANK', 'DEEPAKNTR', 'IRCTC', 'ACC', 'PIDILITIND', 'HINDUNILVR', 'MINDTREE', 'HDFC', 'EICHERMOT', 'ASIANPAINT']
nseTickerList=[]

tempList = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[1].Symbol.to_list()
tickerList = tickerList + tempList
tickerList = list(set(tickerList))

for i in tickerList:
    nseTickerList.append(i+'.NS')

nseTickerList.remove('HDFCLIFE.NS')
nseTickerList.remove('SBICARD.NS')

df = yf.download(nseTickerList, start='2015-01-01' ) 
df = df['Adj Close']

monthlyPrices = df.resample('M').last()
weeklyPrices = df.resample('W').last()
formationDate = dt.today 
```

```python
def monthlyMomentumProfit(formation, monthlyPrices, holdingPeriod=1):
    startFormationDate = formation
    beginMeasurement = startFormationDate - MonthEnd(12)
    endMeasurement = startFormationDate - MonthEnd(1)
    priceEnd = monthlyPrices.loc[endMeasurement]
    price12monthBefore = monthlyPrices.loc[beginMeasurement]
    returnOver12Months =priceEnd / price12monthBefore -1
    winners = returnOver12Months.nlargest(2)
    winReturn = monthlyPrices.loc[startFormationDate + MonthEnd(holdingPeriod),winners.index] / monthlyPrices.loc[startFormationDate,winners.index] - 1
    momentumProfit = winReturn.mean()
    return momentumProfit
```

```python
#As backtest on 2 Jan 2022, 2 Months holding period gave good result as per Excel sheet calculation

def getMonthlyHoldingReturs(monthlyPrices):
    profits, dates, holdingPeriod = [],[],[]

    for i in (1,2,3,4,5,6,9,12):
        for j in monthlyPrices.index[12:-i]:
            profits.append(monthlyMomentumProfit(j,monthlyPrices,holdingPeriod=i))
            dates.append(j+MonthEnd(i))
            holdingPeriod.append(i)

    resultDataFrame = pd.DataFrame({'MomentumProfit': profits, 'HoldingPeriod': holdingPeriod}, index = dates)
    resultDataFrame['MomentumProfit'] = resultDataFrame.MomentumProfit*100
    return resultDataFrame.groupby('HoldingPeriod').mean()
    
```

```python
# Get Monthly Winners for this month

def monthlyMomentumWinner(formation,monthlyPrices, holdingPeriod=3):
    print(f'monthlyMomentumWinner invoked')
    startFormationDate = formation
    beginMeasurement = startFormationDate - MonthEnd(12)
    endMeasurement = startFormationDate - MonthEnd(1)
    print(f'beginMeasurement : {beginMeasurement}, endMeasurement :{endMeasurement}')
    priceEnd = monthlyPrices.loc[endMeasurement]
    price12monthBefore = monthlyPrices.loc[beginMeasurement]
    returnOver12Months = (priceEnd / price12monthBefore -1)*100
    winners = returnOver12Months.nlargest(10)
    return winners

# Get Weekly Winners for this month

def weeklyMomentumWinner(formation, weeklyPrices, holdingPeriod=4,fallBackPeriod=12):
    print(f'weeklyMomentumProfit invoked')
    startFormationDate = formation
    beginMeasurement = startFormationDate - Week(fallBackPeriod)
    endMeasurement = startFormationDate - Week(1)
    print(f'beginMeasurement : {beginMeasurement}, endMeasurement :{endMeasurement}')
    priceEnd = weeklyPrices.loc[endMeasurement]
    price12weekBefore = weeklyPrices.loc[beginMeasurement]
    returnOver12Weeks = (priceEnd / price12weekBefore -1)*100
    winners = returnOver12Weeks.nlargest(10)
    return winners
```

```python
def weeklyMomentumProfit(formation, weeklyPrices, holdingPeriod=1,fallBackPeriod=12):
    startFormationDate = formation
    beginMeasurement = startFormationDate - Week(fallBackPeriod)
    endMeasurement = startFormationDate - Week(1)
    priceEnd = weeklyPrices.loc[endMeasurement]
    price12weekBefore = weeklyPrices.loc[beginMeasurement]
    returnOver12Weeks = priceEnd / price12weekBefore -1
    winners = returnOver12Weeks.nlargest(3)
    winReturn = weeklyPrices.loc[startFormationDate + Week(holdingPeriod),winners.index] / weeklyPrices.loc[startFormationDate,winners.index] - 1
    momentumProfit = winReturn.mean()
    return momentumProfit
```

```python

def getWeeklyHoldingReturs(weeklyPrices,fallBackPeriod=12):
    profits, dates, holdingPeriod = [],[],[]

    for i in (3,4,6,8,10,12):
        for j in weeklyPrices.index[fallBackPeriod:-i]:
            profits.append(weeklyMomentumProfit(j,weeklyPrices,holdingPeriod=i,fallBackPeriod=fallBackPeriod))
            dates.append(j+Week(i))
            holdingPeriod.append(i)

    resultDataFrame = pd.DataFrame({'MomentumProfit': profits, 'WeeklyHoldingPeriod': holdingPeriod}, index = dates)
    resultDataFrame['MomentumProfit'] = resultDataFrame.MomentumProfit * 100
    resultDataFrame['FallBackPeriod'] = fallBackPeriod
    return resultDataFrame.groupby('WeeklyHoldingPeriod').mean()
    
```

```python
#Winner = 1 AND Month=4, 3, 2 gives best result
monthlyPrices = df.resample('M').last()
monthlyReturns = getMonthlyHoldingReturs(monthlyPrices)
monthlyReturns
```

```python
#Get monthly Winner
monthlyPrices = df.resample('M').last()
monthlyReturns = monthlyMomentumWinner(monthlyPrices)
monthlyReturns
```

```python
#Get weekly Winner
weeklyPrices = df.resample('W').last()
formationDate = dt.today 
weeklyReturns = weeklyMomentumWinner(formationDate, weeklyPrices)
weeklyReturns
```

```python
#Lets fetch all the trades taken

def monthlyMomentumTrades(formation,monthlyPrices, holdingPeriod=1):
    order =[]
    
    startFormationDate = formation
    beginMeasurement = startFormationDate - MonthEnd(12)
    endMeasurement = startFormationDate - MonthEnd(1)
    priceEnd = monthlyPrices.loc[endMeasurement]
    price12monthBefore = monthlyPrices.loc[beginMeasurement]
    returnOver12Months =priceEnd / price12monthBefore -1
    winners = returnOver12Months.nlargest(2)
    winReturn = monthlyPrices.loc[startFormationDate + MonthEnd(holdingPeriod),winners.index] / monthlyPrices.loc[startFormationDate,winners.index] - 1
    momentumProfit = winReturn.mean()
    buyDate = startFormationDate
    sellDate= startFormationDate + MonthEnd(holdingPeriod)
    order.append((winners.index[0],returnOver12Months[0],buyDate.to_pydatetime(),monthlyPrices.loc[startFormationDate,winners.index][0],sellDate.to_pydatetime(),monthlyPrices.loc[startFormationDate + MonthEnd(holdingPeriod),winners.index][0],momentumProfit*100,holdingPeriod))
    test = pd.DataFrame(order,columns=['Ticker','PastReturn','BuyDate','BuyPrice','SellDate','SellPrice','Profit','HoldingPeriod'])
    return test


def getMonthlyHoldingTrades(df):
    orders = []
    for i in (1,2,3,4,5,6,9,12):
        for j in monthlyPrices.index[12:-i]:
            orders.append(monthlyMomentumTrades(j,monthlyPrices,holdingPeriod=i))
    return orders

# Run for specific Holding Period
orders = []
for j in monthlyPrices.index[12:-3]:
    orders.append(monthlyMomentumTrades(j,monthlyPrices,holdingPeriod=12))

finalResults = pd.concat(orders)
finalResults.reset_index()

# Run to identify best holding period

#finalResults = pd.concat(getMonthlyHoldingTrades(df))
#finalResults.reset_index()
```
